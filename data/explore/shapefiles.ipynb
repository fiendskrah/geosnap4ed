# NCES shapefiles (California)
downloaded from [this page](https://nces.ed.gov/programs/edge/Geographic/LocaleBoundaries). Available as a single state dataset or the entire US.

```python
data = gpd.read_file('../edge_locale19_nces_CA/edge_locale19_nces_CA.shp')
```

```python
data.shape
```

```python
data
```

### [this pdf](https://nces.ed.gov/programs/edge/docs/LOCALE_DEFINITIONS.pdf) contains definitions for locale classifications.

Four major classifcations: *City*, *Suburban*, *Town*, and *Rural*. Each classifcation has 3 subtypes to further differentiate locales. Subtypes are differentiated by size (in the case of City and Suburban assignments) and proximity (in the case of Town and Rural assignments)

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1)
data.plot(figsize=(12,16),        
          column='LOCALE',
          ax=ax,
          cmap='brg'
          )
```

## We can select specific locales for anlysis.

```python
remote = data.loc[data['LOCALE'] == "43"]
```
